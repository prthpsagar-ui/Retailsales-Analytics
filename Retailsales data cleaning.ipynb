```python
df=pd.read_excel("retail_sales clean.xlsx")
```

```python
print(df)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])
print(df)
```

```python
df['month'] = df['order_date'].dt.month
print(df)
```

```python
df['revenue'] = df['quantity'] * df['unit_price']
print(df)
```

```python
df['revenue'] = df['revenue'] - (df['revenue'] * df['discount_percent']/100)
print(df)
```

```python
sales_by_category = df.groupby('category')['revenue'].sum()
print(sales_by_category)
```

```python
sales_by_city = df.groupby('city')['revenue'].sum()
print(sales_by_city)
```

```python
sales_by_month = df.groupby('month')['revenue'].sum()
print(sales_by_month)
```

```python
plt.figure()
sales_by_month.plot(kind='bar')
plt.title("Monthly Sales")
plt.show()
```
